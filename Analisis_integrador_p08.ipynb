# **Analizando distribuciones: Gráficos de Torta**

Continuando con el proyecto de análisis, procederemos a realizar gráficos de **Torta** que representen dos distribuciones diferentes del dataset **Heart_Faliure** de Hugginface.

Distribuciones de:

- Cantidad de anémicos
- Cantidad de diabéticos
- Cantidad de fumadores
- Cantidad de muertos

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("heart_faliure_limpio.csv")

anemic_count = df['anaemia'].value_counts()
diabetic_count = df['diabetes'].value_counts()
smoker_count = df['smoking'].value_counts()
death_count = df['DEATH_EVENT'].value_counts()

# Definimos las etiquetas para los gráficos de torta
labels = ['No', 'Sí']

# Creamos la figura y los ejes para los subplots con 2 filas y 2 columnas
fig, axs = plt.subplots(1, 4, figsize=(10, 10))

# Añadimos los gráficos de torta
axs[0].pie(anemic_count, labels=labels, autopct='%1.1f%%', startangle=90)
axs[0].set_title('Cantidad de Anémicos')

axs[1].pie(diabetic_count, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1].set_title('Cantidad de Diabéticos')

axs[2].pie(smoker_count, labels=labels, autopct='%1.1f%%', startangle=90)
axs[2].set_title('Cantidad de Fumadores')

axs[3].pie(death_count, labels=labels, autopct='%1.1f%%', startangle=90)
axs[3].set_title('Cantidad de Muertos')

# Ajustamos el layout para evitar superposiciones
plt.tight_layout()

# Mostramos la figura completa
plt.show()
```
