# **Analizando distribuciones: Histogramas**

Continuando con el proyecto de análisis, procederemos a realizar gráficos de **Histograma** que representen dos distribuciones diferentes del dataset **Heart_Faliure** de Hugginface.

1. Distribución de edades:

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("heart_faliure_limpio.csv")

bins = range(int(df['age'].min()), int(df['age'].max()) + 5, 5)

plt.hist(df["age"], bins=bins, edgecolor='black')
plt.title('Distribución de Edades')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')

plt.show()
```

2. Distribución por hombre y mujer según:
- Cantidad de anémicos
- Cantidad de diabéticos
- Cantidad de fumadores
- Cantidad de muertos

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

columnas = ['anaemia', 'diabetes', 'smoking', 'DEATH_EVENT']
categorias = ["Anémicos", "Diabéticos", "Fumadores", "Muertos"]

conteo_hombres = df[df['sex'] == 1][columnas].sum()
conteo_mujeres = df[df['sex'] == 0][columnas].sum()

# Crear el histograma agrupado
ind = np.arange(len(columnas))  
ancho = 0.35  

fig, ax = plt.subplots()
barras_hombres = ax.bar(ind - ancho/2, conteo_hombres, ancho, label='Hombres', color='blue')
barras_mujeres = ax.bar(ind + ancho/2, conteo_mujeres, ancho, label='Mujeres', color='red')

# Etiquetas
ax.set_xlabel('Categorías')
ax.set_ylabel('Cantidad')
ax.set_title('Histograma Agrupado por Sexo')
ax.set_xticks(ind)
ax.set_xticklabels(categorias)
ax.legend()

# Mostrar el gráfico
plt.show()
```
