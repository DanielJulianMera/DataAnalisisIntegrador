# **Prediciendo datos de una columna**

Imagina que tenemos datos faltantes en la colúmna de edades, podríamos usar un modelo para estimar los valores faltantes en base a las otras colúmnas.

Para este laboratorio:

1. Elimina las colúmnas DEATH_EVENT, age y categoria_edad del dataframe para que sea la matriz X
2. Ajusta una regresión lineal sobre el resto de colúmnas y usa la colúmna age como vector y
3. Predice las edades y compara con las edades reales
4. Calcula el error cuadrático medio.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import numpy as np

# Cargar el dataset
file_path = 'heart_faliure_limpio.csv'
data = pd.read_csv(file_path)

# Eliminar columnas DEATH_EVENT, age, y age category para formar la matriz X
X = data.drop(columns=['DEATH_EVENT', 'age', 'age category'])
y = data['age']

# Ajustar un modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)

# Predecir las edades
y_pred = model.predict(X)

# Calcular el MSE
mse = mean_squared_error(y, y_pred)
print(f"Error Cuadrático Medio (MSE): {mse}")

# Calcular el RMSE
rmse = np.sqrt(mse)
print(f"Raíz del Error Cuadrático Medio (RMSE): {rmse}")

# Generar el gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(y, y_pred, alpha=0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--')
plt.xlabel('Edades Reales')
plt.ylabel('Edades Predichas')
plt.title(f'Predicción de Edades (MSE: {mse:.2f}, RMSE: {rmse:.2f})')
plt.grid(True)
plt.show()
```
